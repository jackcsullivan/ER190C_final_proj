# Pull in Observed Solar Radiation Data

```python
def get_df(location_id):
    df = pd.DataFrame()
    for year in range(2006, 2011):
        year_df = pd.read_csv("data/NSRDB_StationData_19910101_20101231_" + str(location_id) +"/NSRDB_StationData_" + str(year) +"0101_" + str(year) +"1231_" + str(location_id) +".csv")
        df = pd.concat([df, year_df])
    return df
```

```python
def make_timestamp_str(row):
    if str(row['HH:MM (LST)']) == '24:00':
        string = str(pd.to_datetime(row['YYYY-MM-DD']) + pd.DateOffset(1))
    else:
        string = str(row['YYYY-MM-DD']) + ' ' + str(row['HH:MM (LST)'])
    return str(pd.to_datetime(string))
```

```python
import os
import pandas as pd
import solar
```

```python
def add_solar(filepath, lat, lon, ts):
    df = pd.read_csv(filepath)
    ts_array = df.apply(make_timestamp_str, axis=1)
    df.index = ts_array
    radiation = solar.radiation_apply_funct(lat, lon, ts)
    result = [radiation(t) for t in ts_array]
    df['clear_sky'] = result
    return df
```

```python
stations = pd.read_csv('stations.csv')
stations
```

```python
os.listdir('./data/solar/')
```

```python
stations['filepath'] = ['li_macarthur_solar.csv', 'newark_solar.csv', 'jfk_solar.csv', 'laguardia_solar.csv','bridgeport_solar.csv']
```

```python
stations
```

```python
dfs = {}
for i in range(len(stations)):
    s = stations.iloc[i]
    print(s['name'])
    result = add_solar('./data/solar/' + s['filepath'], s['lat'], s['lon'], 'US/Eastern')
    dfs[s['name']] = result
    result.to_csv('./data/pysolar/' + s['name']+'.csv')
    
```
